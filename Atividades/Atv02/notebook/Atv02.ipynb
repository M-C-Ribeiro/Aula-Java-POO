# Atv02

Exibir a mensagem “Informe um inteiro: “, deixando o cursor na mesma linha

Atribuir o produto das variáveis b e c para a variável a

Demonstrar a execução do cálculo e utilizar um comentário para afirmar que é um programa de exemplo de folha de pagamento.

```java
import java.util.*; // importando biblioteca

System.out.println("Informe um inteiro: "); // mensagem informativa

Scanner sc = new Scanner(System.in); // criando objeto scanner

int b = sc.nextInt(); // inicializando variável
int c = sc.nextInt(); // inicializando variável

int a = b * c; // atribuindo o produto de b e c, para a variável 
 
System.out.println(b +" x "+ c +" = "+ a); // demonstração do cálculo 
System.out.println("Programa de exemplo de folha de pagamento"); // mensagem informativa
```
